### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,100)
y=x**2
plt.plot(x,y)
plt.title("Square of x (y=x^2)")
plt.xlabel("Square of x (y=x^2)")
plt.xlabel("x values")
plt.ylabel("y values (x^2)")
plt.grid(True)
plt.show()
```
