### Exercise 6: Histogram Customization

```python
import matplotlib.pyplot as plt
data=[1,2,2,3,3,3,4,4,4,4,5]
plt.hist(data,bins=5,color='pink',edgecolor='black')
plt.title("Histogram of Data")
plt.xlabel("value")
plt.ylabel("Frequency")
plt.show()
```
